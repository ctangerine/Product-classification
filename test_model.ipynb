```python
# Load yolo model
from ultralytics import YOLO
```

```python
model = YOLO("best_copy.pt")
```

```python
test_imgs_path = r"E:\DUT Courses\Academic year 4\IOT\Product-classification\Fruit Sugar Detection.v2i.yolov8\test\images"
test_imgs_label_path = r"E:\DUT Courses\Academic year 4\IOT\Product-classification\Fruit Sugar Detection.v2i.yolov8\test\labels"
```

```python
# rename al file in test_imgs_path to number, start from 1 to n
import os
import shutil

for i, img in enumerate(os.listdir(test_imgs_path)):
    shutil.move(os.path.join(test_imgs_path, img), os.path.join(test_imgs_path, f"{i+1}.jpg"))

# rename al file in test_imgs_label_path to number, start from 1 to n
for i, img in enumerate(os.listdir(test_imgs_label_path)):
    shutil.move(os.path.join(test_imgs_label_path, img), os.path.join(test_imgs_label_path, f"{i+1}.txt"))
```

```python
results_test = model("fruit.jpg")  # return a list of Results objects

# Process results list
for result in results_test:
    boxes = result.boxes.cls.cpu()
    print(boxes.numpy()[0])
    print (result.names[2])
```

```python
# predict all images in test_imgs_path and save result
results = {}
for i, img in enumerate(os.listdir(test_imgs_path)):
    img_path = os.path.join(test_imgs_path, img)
    result = model(img_path, conf=0.75)
    for r in result:
        boxes = r.boxes
        results[img] = boxes.cls.tolist()

print (results)
```

```python
print (results)
```

```python
# make a result dict from test_imgs_label_path to compare with results
import numpy as np
results_label = {}
for i, label in enumerate(os.listdir(test_imgs_label_path)):
    with open(os.path.join(test_imgs_label_path, label), 'r') as f:
        print ('file: ', label)
        lines = f.readlines()
        boxes = []
        for line in lines:
            box = line.split()
            box = [float(b) for b in box]
            boxes.append(box)
        boxes = np.array(boxes)
        print ('box:', boxes.shape)
        results_label[label] = boxes[:, 0].tolist()

print (results_label)
```

```python
#convert key of results_label from 'number.txt' to 'number.jpg'
results_label = {f"{k.split('.')[0]}.jpg": v for k, v in results_label.items()}
```

```python
# compare results and results_label
# count sum of match label in each image
count = 0
count_per_image = 0
sum_label = 0
accuracy_stat = {}
for key, values in results.items():
    count_per_image = 0
    for v in values:
        if v in results_label[key]:
            count += 1
            count_per_image += 1
    accuracy_per_image = count_per_image*100/len(results_label[key])
    #print (f"Accuracy of {key}: {accuracy_per_image}%")
    
    accuracy_stat[key] = accuracy_per_image

for key, values in results_label.items():
    sum_label += len(values)

print (f"Total accuracy: {count*100/sum_label}%")

```

```python
# calculate stat of accuracy_stat
accuracy_stat_org = accuracy_stat.copy()
import numpy as np
accuracy_stat = np.array(list(accuracy_stat_org.values()))
print (f"Mean accuracy: {accuracy_stat.mean()}")
print (f"Max accuracy: {accuracy_stat.max()}")
print (f"Min accuracy: {accuracy_stat.min()}")
print (f"Std accuracy: {accuracy_stat.std()}")
print (f"Median accuracy: {np.median(accuracy_stat)}")

# print number per each percentage
print (f"Number of image with accuracy >= 95%: {np.sum(accuracy_stat >= 95)} on {len(accuracy_stat)}")
print (f"Number of image with accuracy >= 90%: {np.sum(accuracy_stat >= 90)} on {len(accuracy_stat)}")
print (f"Number of image with accuracy >= 85%: {np.sum(accuracy_stat >= 85)} on {len(accuracy_stat)}")
print (f"Number of image with accuracy >= 80%: {np.sum(accuracy_stat >= 80)} on {len(accuracy_stat)}")
print (f"Number of image with accuracy <= 50%: {np.sum(accuracy_stat <= 50)} on {len(accuracy_stat)}")
print (f"Number of image with accuracy <= 20%: {np.sum(accuracy_stat <= 20)} on {len(accuracy_stat)}")

#list of image with accuracy <= 50%
imgs = []
for key, value in accuracy_stat_org.items():
    if value <= 50:
        print (f"Image {key} with accuracy {value}%, label: {results_label[key]} and predict: {results[key]}")
        imgs.append(key)
```

```python
print (imgs)
```

```python
#results_test = model((f"{test_imgs_path}/682.jpg"))  # return a list of Results objects
underscore_result = {}
underscore_path = []
# add test_imgs_path to each path in imgs
for img in imgs:
    img = os.path.join(test_imgs_path, img)
    underscore_path.append(img)

index = 0
for img in underscore_path:
    result = model(img, conf=0.2)
    for r in result:
        boxes = r.boxes
        underscore_result[imgs[index]] = boxes.cls.tolist()
        index += 1

print (underscore_result)


```

```python
# calculate the same
count_underscore = 0
count_per_image_underscore = 0
sum_label_underscore = 0
accuracy_stat_underscore = {}
for key, values in underscore_result.items():
    count_per_image_underscore = 0
    for v in values:
        if v in results_label[key]:
            count_underscore += 1
            count_per_image_underscore += 1
    accuracy_per_image = count_per_image_underscore*100/len(results_label[key])
    #print (f"Accuracy of {key}: {accuracy_per_image}%")
    
    accuracy_stat_underscore[key] = accuracy_per_image

for key, values in results_label.items():
    if key in underscore_result.keys():
        sum_label_underscore += len(values)

print (f"Total accuracy: {count_underscore*100/sum_label_underscore}%")
```

```python
accuracy_stat_underscore_org = accuracy_stat_underscore.copy()
import numpy as np
accuracy_stat_underscore = np.array(list(accuracy_stat_underscore_org.values()))
print (f"Mean accuracy: {accuracy_stat_underscore.mean()}")
print (f"Max accuracy: {accuracy_stat_underscore.max()}")
print (f"Min accuracy: {accuracy_stat_underscore.min()}")
print (f"Std accuracy: {accuracy_stat_underscore.std()}")
print (f"Median accuracy: {np.median(accuracy_stat_underscore)}")

# print number per each percentage
print (f"Number of image with accuracy >= 95%: {np.sum(accuracy_stat_underscore >= 95)} on {len(accuracy_stat_underscore)}")
print (f"Number of image with accuracy >= 90%: {np.sum(accuracy_stat_underscore >= 90)} on {len(accuracy_stat_underscore)}")
print (f"Number of image with accuracy >= 85%: {np.sum(accuracy_stat_underscore >= 85)} on {len(accuracy_stat_underscore)}")
print (f"Number of image with accuracy >= 80%: {np.sum(accuracy_stat_underscore >= 80)} on {len(accuracy_stat_underscore)}")
print (f"Number of image with accuracy <= 50%: {np.sum(accuracy_stat_underscore <= 50)} on {len(accuracy_stat_underscore)}")
print (f"Number of image with accuracy <= 20%: {np.sum(accuracy_stat_underscore <= 20)} on {len(accuracy_stat_underscore)}")

#list of image with accuracy <= 50%
imgs_new = []
for key, value in accuracy_stat_underscore_org.items():
    if value <= 30:
        print (f"Image {key} with accuracy {value}%, label: {results_label[key]} and predict: {underscore_result[key]}")
        imgs_new.append(key)

print(imgs_new)
```
